Pandas ignores missing values by default so if there are not many missing values we can simply ignore them. But if there are a lot of values or for any other reason if we want to delete missing values or replace them with some other logical values we can always do so.

```python
titanic.isna()
#isnull()
```

```python
#titanic.notna()
titanic.notna().sum()
```

```python
titanic.dropna(how='all').shape
```

```python
titanic.dropna(how='any', axis=1).shape
```

```python
#titanic.dropna(how='any', axis=1,thresh=600).shape
```

```python
titanic.dropna(axis=1,thresh=600, inplace=True)
titanic.head()
```

```python
titanic.loc[titanic.Age.isnull()]
```

```python
avg = titanic.Age.mean()
avg
```

```python
titanic.Age.fillna(value=avg, inplace=True)
titanic.loc[titanic.Age.isna()]
```

```python
data = {
    'Open': [100, 105, np.nan, 108, np.nan, 112],
    'High': [102, 108, np.nan, 110, np.nan, 115],
    'Low': [98, 103, np.nan, 106, np.nan, 110],
    'Close': [101, 107, np.nan, 109, np.nan, 113],
}
df = pd.DataFrame(data)
df
```

```python
# Fill missing values using different methods
df.fillna(method='ffill')  # Forward fill
```

```python
df.fillna(method='bfill')  # Backward fill
```

```python
df.interpolate()  # Linear interpolation
```

```python
df.replace({1000: 10, 2000: 60})
```
